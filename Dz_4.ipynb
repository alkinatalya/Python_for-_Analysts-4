### Задача 1.1

```python
import pandas as pd
import matplotlib.pyplot as plt
#import seaborn as sns


df = pd.read_csv('kc_house_data.csv')
df.head()
```

### Задача 1.2

```python
plt.hist(df['price'])
plt.title('Распределение цены')
plt.xlabel('Цена')
plt.ylabel('Количество');
```

### Задача 1.3

```python
plt.hist(df['sqft_living'])
plt.title('Распределение жилой квадратуры')
plt.xlabel('Жилая квадратура')
plt.ylabel('Количество');
```

Задача 1.4(2.1.4)

Задача 3

```python
import seaborn as sns
sns.histplot(df['price'])
plt.xlabel('price')
plt.ylabel('count')
plt.title('Cтоимость недвижимости ')

```

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Влияние оценки на стоимость недвижимости ')
```

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Влияние этожей на стоимость недвижимости ')
```

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Влияние вида на стоимость недвижимости ')
```

```python
sns.boxplot(x=df['price'], y=df['bedrooms'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('bedrooms')
plt.title('Влияние комнат на стоимость недвижимости ')
```
